pipeline


```python
from transformers import pipeline
```

```python
classifier = pipeline ('sentiment-analysis')
```

```python
classifier("I am very happy to learn python  and its applications")
```

Mutiple sentance

```python
sentences = ["I hate football", "I love to play carrom"]
classifier(sentences)
```

Zero-shot classification

```python
classifier = pipeline ('zero-shot-classification' , model='MoritzLaurer/DeBERTa-v3-base-mnli-fever-anli')
```

```python

classifier("This is about Machine Learning and natural language processing",
           candidate_labels = ['Education', 'Sports', 'Business'])
```

Text Genaration

```python
classifier = pipeline('text-generation', model='distilgpt2')
```

```python
classifier("In this Generative AI course, we will teach you how to do the process")
```

NER- Named Enitity recognition

Unwrap the pipeline

1. Preprocessing the text

```python
# understanding the tokenization task
from transformers import AutoTokenizer
model = "mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis"
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
# Trail of this tokenizer
FullInput = [" I am happy to learn new things which is python", "The python is very tough to learn "]

Fullinputs = tokenizer(FullInput, padding=True, truncation=True, return_tensors="pt")
```

```python
print(Fullinputs)
```

Feeding the input to the model

```python
from transformers import AutoModel
model_checkpoint = "mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis"
model = AutoModel.from_pretrained(model_checkpoint)
```

```python
outputs = model(**Fullinputs)
print(outputs)
```

```python
from transformers import AutoModelForSequenceClassification
model_checkpoint = "mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis"
model = AutoModelForSequenceClassification.from_pretrained(model_checkpoint)
```

```python
outputs = model(**Fullinputs)
print(outputs.logits)
```

```python
import torch
torch.nn.functional.softmax(outputs.logits)
```
