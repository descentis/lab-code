```python
from transformers import pipeline
```

```python
classifier = pipeline('sentiment-analysis')
```

```python
classifier('I love my family')
```

```python
sentances = ["I love work from home culture", "I hate go to office"]
classifier(sentances)
```

```python
classifier = pipeline('zero-shot-classification',model="facebook/bart-large-mnli")
```

```python
classifier("This cource for deeplearning, LLM, M/c learning and datascience",
candidate_labels=["business","training","education","college"])
```

```python
classifier = pipeline('text-generation',model="distilgpt2")
```

```python
classifier("I want to learn LLM")
```

```python
classifier = pipeline("ner")
```

```python
sentences=["my DOB 18 july 1989","my office in cbe"]
classifier(sentences)
```
