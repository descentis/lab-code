```python
from transformers import pipeline
```

```python
classifier = pipeline('sentiment-analysis')
```

```python
classifier("I watched a feel good movie")
```

```python
sentences = ["I hate coffee", "I love football"]
classifier(sentences)
```

```python
classifier = pipeline('zero-shot-classification', model='MoritzLaurer/bge-m3-zeroshot-v2.0')
```

```python
classifier("I love to eat food",
           candidate_labels = ['food','sports'])
```

```python
classifier = pipeline('text-generation', model='openai-community/gpt2')
```

```python
classifier("lets make coffee, the steps are")
```

NER - named entity recognition

```python
classifier = pipeline('ner', model='Jean-Baptiste/camembert-ner')
```

```python
sentences = ["My name is Swathi"]
classifier(sentences)
```

```python
from transformers import AutoTokenizer
model = "Jean-Baptiste/camembert-ner"
tokenizer=AutoTokenizer.from_pretrained(model)
input = ["I am from chennai", "I know tamil"]

tokenizer(input)
```
