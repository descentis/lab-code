#pipeline


```python
from transformers import pipeline
```

```python
classifier=pipeline('sentiment-analysis')
```

```python
classifier("i have to go to school tommorow")
```

#if we have multiple sentences

```python
sentences=["i have work tomorrow","i dont need to go to office"]
classifier(sentences)
```

# zero-short classification

```python
from transformers import pipeline
```

```python
classifier = pipeline('zero-shot-classification')
```

```python
classifier("I love to travel because it allows me to step outside of my everyday routine and immerse myself in new environments, cultures, and experiences. Traveling gives me the chance to meet diverse people, explore breathtaking landscapes, and learn about the history and traditions of different places. It fuels my curiosity and sense of adventure, broadening my perspective and leaving me with memories that last a lifetime. Each journey offers personal growth, new insights, and a deeper appreciation for the world around me.",candidate_labels=['journey','environments'])
```
