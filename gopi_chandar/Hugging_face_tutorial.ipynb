# Unwrap the pipeline

1.   Preprocessing the text




```python
# Let's understand the tokenization task
from transformers import AutoTokenizer
model="distilbert/distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(model)

```

```python
# Working of Tokenizer
raw_input = ["I am very excited to learn transformers and its applications", "The instructor is really bad"]
tokens = tokenizer(raw_input, padding=True, truncation=True, return_tensors="pt")
print(tokens)
```

Feeding the input to the Model

```python
from transformers import AutoModel
model_checkpoint="distilbert/distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModel.from_pretrained(model_checkpoint)
output = model(**tokens)
# print(output.last_hidden_state.shape)
print(output)
```

```python
from transformers import AutoModelForSequenceClassification
model_checkpoint="distilbert/distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(model_checkpoint)
output = model(**tokens)
print(output.logits)
```

```python
import torch
predictions = torch.nn.functional.softmax(output.logits)
print(predictions)
```
