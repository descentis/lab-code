#Pre processing the text

```python
from transformers import AutoTokenizer
model = "distilbert/distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
input = ["THis is test sentence",
         "Hope this will work good"]
tokens = tokenizer(input, padding=True, truncation=True, return_tensors='pt')
print(tokens)
```

```python
from transformers import AutoModel
model_cp = "distilbert/distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModel.from_pretrained(model_cp)
output = model(**tokens)
print(output)
```

```python
from transformers import AutoModelForSequenceClassification
model_checkpoint = "distilbert/distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(model_checkpoint)
output = model(**tokens)
print(output.logits)
```

```python
import torch
torch.nn.functional.softmax(output.logits)
```
